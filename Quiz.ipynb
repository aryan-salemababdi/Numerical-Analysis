```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
def f(x):
    return x**2

def approximate_area_under_curve(num_points):
    points_under_curve = []
    for _ in range(num_points):
        x_point = np.random.uniform(0, 2)
        y_point = np.random.uniform(0, 4)
    
        if y_point <= f(x_point):
            points_under_curve.append((x_point, y_point))
    total_points = num_points
    area_under_curve = len(points_under_curve) / total_points
    total_area = area_under_curve * 8
    
    return total_area, np.array(points_under_curve), (num_points-len(points_under_curve)), (8-total_area)

num_points = 100000
approximated_area, points_under_curve, num_outer_curve, approximated_outer_area = approximate_area_under_curve(num_points)
print(f"total area under curve:{approximated_area}, points_under_curve:{len(points_under_curve)},  outer:{num_outer_curve} and outer area is:  {approximated_outer_area}")
x = np.linspace(0, 2, 10000)
y = f(x)
plt.plot(x, y, 'r', label='y = x^2')
plt.scatter(points_under_curve[:, 0], points_under_curve[:, 1], color='blue', s=5, label='Under Curve')
plt.text(0.5, 0.2, f"Approximated Area: {approximated_area:.4f}", ha='center')
currentAxis = plt.gca()
currentAxis.add_patch(patches.Rectangle((0, 0), 2, 2, linewidth=1, edgecolor='green', facecolor='none', label='2x2 Square'))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area under Curve y = x^2')
plt.legend()
plt.grid(True)
plt.show()
```
