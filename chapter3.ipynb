```python
import numpy as np
import random as rn
import matplotlib.pyplot as plt
```

**Plot Functions**

```python
def func_d3(x):
    return x**3 - 3 * x**2 - 12 * x + 10
def func_d2(x):
    return  x**2 + 5 * x
x_vals = np.linspace(-20, 20, 100)
y_vals = func_d3(x_vals)
y_vals2 = func_d2(x_vals)
plt.figure(figsize=(10, 10))
plt.plot(x_vals, y_vals, label="degree3 function", c='b', linewidth=2)
plt.plot(x_vals, y_vals2, label="degree2 function", c='r', linestyle='dashed', linewidth=2)
plt.title("functions")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
```

**Random Numbers and histogram**

```python
random_numbers = [rn.random() for _ in range(100)]
bins = [0] * 10
for number in random_numbers:
    index = int(number * 10)
    bins[index] += 1
for i, bin_count in enumerate(bins):
    bin_range = f"{i / 10:.1f} - {(i + 1) / 10:.1f}"
    print(f"{bin_range}: {'*' * bin_count} - {bin_count}")
print(bins)
plt.bar([i / 10 for i in range(10)], bins, width=0.1, edgecolor='y')
plt.title('Histogram of Random Numbers between 0 and 1')
plt.xlabel('Bins')
plt.ylabel("N")
plt.xticks([i / 10 for i in range(10)])
plt.show()
```

**Insect life analysis**

```python
def insect_survival(time_stamp):
    place = 1
    places = []
    for _ in range(time_stamp):
        x = rn.choice([-1, 1])
        if x == 1:
            place = place + 1
            places.append(place)
        else:
            if place > 0:
                place = place - 1
                places.append(place)
            else:
                place = 0
                places.append(place)
        if place == 0:
            break
    if place == 0:
        return f"Unfortunately, the insect fell off in {len(places)} moves and {time_stamp}s ! It couldn't survive. :((((((((( ",places
    else:
        return f"The insect survived! It's still alive after {time_stamp} and {len(places)} moves! :))))))) ",places
counter = int(input("number of moves : "))
result, places = insect_survival(counter)
print(result, places)
```

**Monte Carlo method**

```python
def estimate_pi(num_samples):
    points_inside_circle = 0
    points_total = num_samples
    inside_circle_x = []
    inside_circle_y = []
    outside_circle_x = []
    outside_circle_y = []
    for _ in range(num_samples):
        x = rn.uniform(-1, 1)
        y = rn.uniform(-1, 1)
        distance = x ** 2 + y ** 2
        if distance <= 1:
            points_inside_circle += 1
            inside_circle_x.append(x)
            inside_circle_y.append(y)
        else:
            outside_circle_x.append(x)
            outside_circle_y.append(y)
    pi_estimate = 4 * (points_inside_circle / points_total)
    return pi_estimate, inside_circle_x, inside_circle_y, outside_circle_x, outside_circle_y
num_samples = 100000
estimated_pi, inside_x, inside_y, outside_x, outside_y = estimate_pi(num_samples)
circle = plt.Circle((0, 0), 1, color='blue', fill=False, label='Circle')
fig, ax = plt.subplots()
ax.add_artist(circle)
plt.scatter(inside_x, inside_y, c='b', label='Inside Circle')
plt.scatter(outside_x, outside_y, c='r', label='Outside Circle')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Estimation of Pi: {estimated_pi}')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
def func(x):
    return (x**3) + (x**3)/(np.sqrt(x+1))
def monte_carlo(a,b,n):
    h = (b-a)/len(n)
    integral = 0
    for i in range(len(n)):
        integral+=func(n[i])
    return integral * h
n = np.random.uniform(0,5,1000000)
print(monte_carlo(0,5,n))
```
