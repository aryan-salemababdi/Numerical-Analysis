```python
import numpy as np
import matplotlib.pyplot as plt
import random as rn
```

```python
# باز کردن فایل و خواندن داده‌ها
with open('data_x.txt', 'r') as f:
    df = [float(line.strip()) for line in f.readlines()]
bins = 10
n, bins, patches = plt.hist(df, bins=bins, density=True, alpha=0.8)
bin_centers = 0.5 * (bins[1:] + bins[:-1])
pdf = np.histogram(df, bins=bins, density=False)[0]
plt.plot(bin_centers, pdf, 'b', linewidth=2)
plt.show()
```

```python
mean = np.mean(df)
max_df = max(df)
print("میانگین داده‌ها: ", mean)
print(f"maximum is : {max_df}")
```

```python
# محاسبه انحراف معیار داده ها
# std_dev = math.sqrt(sum([(x - mean) ** 2 for x in data]) / (len(data) - 1)) فرمول مربوطه
std_dev = np.std(df)
print(std_dev)
plt.hist(df, bins=10, edgecolor='black')
plt.axvline(mean, color="yellow", linestyle='dashed', linewidth=2)
plt.axvline(mean + std_dev, color="red", linestyle='dashed', linewidth=2)
plt.axvline(mean - std_dev, color="blue", linestyle='dashed', linewidth=2)
plt.show()
```

```python
counter = 0

std_mean_plus = std_dev + mean

std_mean_mines = mean - std_dev

for i in range (len(df)) :
    if df[i] < std_mean_plus and df[i] > std_mean_mines:
        counter+=1
    else:
        pass
print(f"درصد داده ها{(counter/len(df))*100}")    
```
